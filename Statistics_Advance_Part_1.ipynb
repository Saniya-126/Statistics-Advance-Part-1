1. What is a random variable in probability theory?
* A random variable is a numerical outcome of a random phenomenon, mapping events from a probability space to real numbers.

2. What are the types of random variables?
*  The two main types of random variables are discrete random variables, which take on a countable number of values, and continuous random variables, which can take on any value within a given range.

3. What is the difference between discrete and continuous distributions?
* Discrete distributions describe random variables that can take on a finite or countably infinite set of distinct values, such as the number of heads in a series of coin flips. In contrast, continuous distributions represent random variables that can take on any value within a continuous range, such as the height of individuals, and are characterized by probability density functions rather than probability mass functions.

4. What are probability distribution functions (PDF)?
* A probability distribution function (PDF) describes the likelihood of a random variable taking on specific values, with discrete random variables using probability mass functions (PMFs) and continuous random variables using probability density functions.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
* Cumulative distribution functions (CDF) represent the probability that a random variable takes on a value less than or equal to a specific point, while probability distribution functions (PDF) describe the likelihood of the random variable taking on exact values.

6. What is a discrete uniform distribution?
* A discrete uniform distribution is a probability distribution where a finite number of outcomes are equally likely, meaning each outcome has the same probability of occurring.

7. What are the key properties of a Bernoulli distribution?
* The key properties of a Bernoulli distribution include having two possible outcomes (success and failure), a single trial, and a probability (p) of success, with the mean equal to (p) and the variance equal to (p(1-p)).

8. What is the binomial distribution, and how is it used in probability?
* The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success, and is used to calculate the likelihood of obtaining a specific number of successes in those trials.

9. What is the Poisson distribution and where is it applied?
* The Poisson distribution models the number of events occurring in a fixed interval of time or space, given that these events happen with a known constant mean rate and independently of the time since the last event, and is commonly applied in fields such as telecommunications, traffic flow, and queuing theory.

10. What is a continuous uniform distribution?
*  A continuous uniform distribution is a probability distribution where all outcomes within a specified interval are equally likely, characterized by a constant probability density function over that interval.

11. What are the characteristics of a normal distribution?
* The characteristics of a normal distribution include a symmetric bell-shaped curve, defined by its mean (which determines the center) and standard deviation (which determines the spread), with approximately 68% of the data falling within one standard deviation of the mean.

12. What is the standard normal distribution, and why is it important?
* The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1, and it is important because it allows for the standardization of scores, enabling comparisons across different normal distributions using z-scores.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
* The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population's distribution, making it critical for inferential statistics and hypothesis testing.

14. How does the Central Limit Theorem relate to the normal distribution?
*  The Central Limit Theorem relates to the normal distribution by stating that as the sample size increases, the distribution of the sample means will approximate a normal distribution, even if the original population distribution is not normal.

15. What is the application of Z statistics in hypothesis testing?
*  Z statistics are used in hypothesis testing to determine how many standard deviations a sample mean is from the population mean, allowing researchers to assess the significance of their results against a standard normal distribution.

16. How do you calculate a Z-score, and what does it represent?
* A Z-score is calculated by subtracting the population mean from the individual data point and then dividing by the population standard deviation, representing the number of standard deviations a data point is from the mean.

17. What are point estimates and interval estimates in statistics?
* Point estimates provide a single value estimate of a population parameter, while interval estimates offer a range of values within which the parameter is expected to lie, typically accompanied by a confidence level.

18. What is the significance of confidence intervals in statistical analysis?
* Confidence intervals are significant in statistical analysis as they provide a range of values that likely contain the true population parameter, allowing researchers to quantify the uncertainty and reliability of their estimates.

19. What is the relationship between a Z-score and a confidence interval?
* The Z-score is used to determine the critical values that define the boundaries of a confidence interval, indicating how many standard deviations away from the mean the interval extends for a specified confidence level.

20. How are Z-scores used to compare different distributions?
* Z-scores are used to compare different distributions by standardizing values from different datasets to a common scale, allowing for the assessment of relative positions and probabilities regardless of the original distribution's mean and standard deviation.

21. What are the assumptions for applying the Central Limit Theorem?
* The assumptions for applying the Central Limit Theorem include that the samples are independent, the sample size is sufficiently large (typically n â‰¥ 30), and the population from which the samples are drawn has a finite mean and variance.

22. What is the concept of expected value in a probability distribution?
* The expected value in a probability distribution is the long-term average or mean of all possible outcomes, weighted by their probabilities, representing the central tendency of the distribution.

23. How does a probability distribution relate to the expected outcome of a random variable?
* A probability distribution describes the likelihood of all possible outcomes of a random variable, and the expected outcome, or expected value, is calculated as the weighted average of these outcomes, reflecting the average result one would anticipate over many trials.

```python
# Question 3

import numpy as np

def bernoulli_pdf(p):
    """Calculate the PDF of a Bernoulli distribution.

    Args:
        p (float): Probability of success (0 <= p <= 1).

    Returns:
        dict: A dictionary with the probabilities for outcomes 0 and 1.
    """
    if p < 0 or p > 1:
        raise ValueError("Probability p must be between 0 and 1.")

    pdf = {
        0: 1 - p,
        1: p
    }

    return pdf

p = 0.7
pdf = bernoulli_pdf(p)
print(f"PDF of Bernoulli Distribution (p={p}): {pdf}")
```

```python
# Question 6

import numpy as np
import matplotlib.pyplot as plt

low = 1
high = 10
x = np.arange(low, high + 1)
probabilities = np.ones_like(x) / (high - low + 1)

cdf = np.cumsum(probabilities)

plt.figure(figsize=(10, 6))
plt.step(x, cdf, where='post', color='blue', label='CDF', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid()
plt.axhline(1, color='red', linestyle='--', label='Cumulative Probability = 1')
plt.legend()
plt.show()
```

```python
# Question 9

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """Calculate Z-scores for a given dataset."""
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data):
    """Plot the original data and their Z-scores."""
    z_scores = calculate_z_scores(data)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(data, marker='o', linestyle='-', color='blue')
    plt.title('Original Data')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(z_scores, marker='o', linestyle='-', color='orange')
    plt.title('Z-Scores')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.axhline(0, color='red', linestyle='--', label='Z=0')
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()

data = np.random.normal(loc=50, scale=10, size=100)
plot_z_scores(data)
```

```python
# Question 19

import numpy as np
import scipy.stats as stats

population_mean = 100
sample_size = 30
sample_data = np.random.normal(loc=105, scale=10, size=sample_size)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

```python
# Question 20

import numpy as np
import scipy.stats as stats

np.random.seed(0)
data = np.random.normal(loc=170, scale=10, size=100)

sample_mean = np.mean(data)
standard_error = stats.sem(data)

confidence_level = 0.95
alpha = 1 - confidence_level

ci_lower, ci_upper = stats.t.interval(confidence_level, df=len(data)-1, loc=sample_mean, scale=standard_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

print(f"\nInterpretation: We are {confidence_level*100:.0f}% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

```python
# Question 28

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(size=1000, mean=0, std_dev=1):
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    skewness = skew(data)
    kurt = kurtosis(data)

    return skewness, kurt

skewness, kurtosis_value = calculate_skewness_kurtosis()
print(f"Skewness: {skewness}, Kurtosis: {kurtosis_value}")
```
